# Agentic AI
- It is a type of AI that can take up a task or goal from a user & then work toward completing it on its own, wth minimal human guidance.
- It plans, takes, adapts to changes, and seeks help only when necessary

## Characterstics of Agentic AI:

### 1. Autonomy
- It refers to the AI system's ability to make decisions on its own to achieve a given goal, without needing step-by-step human instructions.
- It also means proactive i.e. adapts the situation and handle it accordingly.

- Autonomy is implemented in multiple steps:
    - Execution: Automatically executing all the steps
    - Decision Making: Making decisions automatically
    - Tool Usage: Use tools according to the situation

- Autonomy can be Controlled:
    1. Permission Scope: Limit what tools or actions the agent can perform indepently. (Decisions like updating & deleting a record in a database needs approval)
    2. Human In The Loop: Inset checkpoints where human approval is required before continuing. (Sending Revenue Reports to Stakeholders)
    3. Override Controls: Allow users to stop, pause, or change the agent's behaviour at any time
    4. Guardrails/ Policies: Define hard rules or ethical boundaries the agentmust follow.

### 2. Goal Oriented
- It means that the AI system operates with a persistent objective in mind and continously directs its actions to achieve that objective, rather than just responding to isolated prompts

- Notes:
    1. Goals acts as a compass for Autonomy.
    2. Goal can come with constraints.
    3. Goals are stored in core memory.
    
    4. Goals can be altered.

### 3. Planning
- It's the agent's ability to breakdown a high-level goal into a structured sequence of actions or subgoals and decide the best path to achieve the desired outcome. 
- Example: Hiring a Backend Engineer

- Steps:
    1. Generating multiple plans:
        - Plan A: Post JD on Linkedin, Github Jobs, Naukri.com
        - Plan B: Use internal referrals & hiring agencies

    2. Evaluate each plan:
        - Efficiency (which is faster)
        - Tool Availability (Which tools are available)
        - Cost
        - Risk
        - Alignment (Does the plan align with the goal. Does it met all the condition & constraints)

    3. Select the best plan with the help of:
        - Huma In The Loop: Input, Follow-Up Questions, Approval for specific persons
        - A pre-programmed Policy

### 4. Reasoning
- It is the cognitive process through which an agentic ai system interprets information, draws conclusions, & make decisions while planning & executing actions in real time.

- Reasoning during planning:
    1. Goal Decomposition: Break down goals into concrete steps
    2. Tools Selection
    3. Resource Estimation

- Reasoning During Execution:
    1. Decision Making
    2. HITL Handling
    3. Error Handling

### 5. Adaptability
- It is the agent's ability to modify its plans, strategies, or actions in response to unexpected conditions - all while staying aligned with the goal.
- It helps when:
    1. Faliures
    2. External Feedback
    3. Changing Goals (Hiring a Freelancer instead of Full Time Employee)

### 6. Context Awarness
- It is the agent's ability to understand, retain, & utilize relevant information from the ongoing task, past interactions, user preferences, & environmental cues to make better decisions throughout a multi-step process.

- Contexts: The Original Goal + (Progress Till now + Interaction History) + Tool Responses + Environment State + User Specific preference + Policy or Guardrails

- Context awarness is implemented through memory (Short Term  & Long Term Memory)


## Core Components of an Agentic System
1. Brain (LLM) -> Goal Interpretation, Planning, Reasoning, Tool Selection & Conversation 
2. Orchestrator -> Task Sequencing, Routing, Retry Logic, Delegation, Looping & Iteration
3. Tools -> External Actions & Knowledge Base Access
4. Memory -> Short Term, Long Term & State Tracking (Agent State)
5. Supervisor -> Approval (HITL), Guardrails, Edge Cases
