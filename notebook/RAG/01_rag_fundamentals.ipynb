# RAG:
RAG = Retrieval + LLM Generation  

- Instead of relying only on an LLM’s internal knowledge, you:
    1. Retrieve relevant data from your own knowledge base
    2. Augment the prompt with that data
    3. Let the LLM generate a grounded answer

### Why RAG ?  
- LLMs have limitations:  
    ❌ Hallucinations  
    ❌ Outdated knowledge  
    ❌ No access to private/company data  

- RAG solves this by:  
    ✅ Injecting real, up-to-date data  
    ✅ Making answers explainable (via sources)  
    ✅ Allowing domain-specific knowledge  

### Core Components:
1. Document Loaders
2. Text Splitters
3. Embeddings
4. Vector Stores
5. Retriever
6. LLM
7. Prompt
