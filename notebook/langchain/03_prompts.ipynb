# Prompts
Prompts are input instructions or queries given to a model to guide its output.

## Type of Prompts:
1. Text-based prompts
    - Static Prompts
    - Dynamic Prompts
2. Multimodal prompts

## Messages in Langchain:
For building chatbots, we need to maintain conversation history between the user and the AI. Instead of manually managing this (e.g., using dictionaries), LangChain provides built-in message types that simplify the process:
1. SystemMessage – sets behavior/instructions for the AI
2. HumanMessage – represents user input
3. AIMessage – stores the AI’s response  

These structured messages make it easier to manage context and build conversational flows.

### A very simple Conversational Chatbot with History

```python
from langchain_groq import ChatGroq
from langchain_core.messages import SystemMessage, HumanMessage, AIMessage
from dotenv import load_dotenv

load_dotenv()

model = ChatGroq(
    model="openai/gpt-oss-safeguard-20b",
    temperature=0.7,
)

chat_history = [SystemMessage(content='You are a helpful AI assistant')]
while True:
    user_input = input('You: ')
    chat_history.append(HumanMessage(content=user_input))
    if user_input == 'exit':
        break
    result = model.invoke(chat_history)
    chat_history.append(AIMessage(content=result.content))
    print("AI: ",result.content)

print("---" * 10, "Chat History", "---" * 10)
print(chat_history)
```

## PromptTemplate
- A promptTemplate in langchain is a structured way yo create prompts dynamically by inserting variables into a predefined template.
- Instead of hardcoding prompts, promptTemplate allows you to define placeholders that can be filled in at runtime with different inputs.
- This makes it reusable, flexible, & easy to manage, especially when working with dynamic user inputs or automated workflow.
- A simple template for generating a single prompt string (non-chat models or basic use cases).

### When to use PromptTemplate:
1. You are working with LLMs (text-in, text-out)
2. No multi-turn conversation needed
3. You just need one formatted prompt

```python
from langchain_core.prompts import PromptTemplate

template = PromptTemplate.from_template(
    "Write a summary about {topic} in {tone} tone."
)

prompt = template.format(topic="AI", tone="simple")
```

## ChatPromptTemplate
A structured prompt made of multiple messages (System, Human, AI) for chat models.

### When to use ChatPromptTemplate:
1. You are working with chat models
2. Need conversation structure
3. Want to define roles (system, user, assistant)

### Why use template over f stings ?
1. Default Validation: Give error if any dynamic variable is not present or something etc is given.
2. Reusable: We can store prompts as JSON or different formats & can be used anywhere.
3. Langchain Ecosystem

```python
from langchain_core.prompts import ChatPromptTemplate

template = ChatPromptTemplate.from_messages([
    ("system", "You are a helpful assistant."),
    ("human", "Explain {topic} in simple terms")
])

prompt = template.format_messages(topic="AI")
```

## Message Placeholders
A message placehold  er is a special placeholder used inside a ChatPromptTemplate to dynamically insert chat history or a list of messages at run time.

### When to Use:
When building chatbots, you often have previous conversation history. Instead of manually adding all past messages, MessagesPlaceholder lets you inject them automatically.

```python
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder
chat_template = ChatPromptTemplate([
    ('system','You are a helpful customer support agent'),
    MessagesPlaceholder(variable_name='chat_history'),
    ('human','{query}')
])

chat_history=[
    HumanMessage(content="I want to request a refund for my order #12345."),
    AIMessage(content="Your refund request for order #12345 has been initiated. It will be processed in 3-5 business days.")
]

print(chat_history)

prompt = chat_template.invoke({'chat_history':chat_history, 'query':'Where is my refund'})
print(prompt)
```
